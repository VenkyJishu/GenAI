```python
import re
words =[]
with open('../pdf_files/final.txt','r') as f:
    file_data = f.read()
    file_data = file_data.lower()
    words = re.findall(r'\w+',file_data)

#Vocabulary
vocab = set(words)
len(vocab)


```

```python
def cnt_words(words):
    dict={}
    for word in words:
        if word in dict:
            dict[word] = dict.get(word) + 1
        else:
            dict[word] = 1
    return dict

```

```python
def check_probability_fr_word(word_dict):
    prob ={}
    m = sum(word_dict.values())
    print(f' m is {m}')

    for key in word_dict.keys():
        prob[key] = word_dict[key] / m
    return prob

word_dict = cnt_words(words=words)
check_probability_fr_word(word_dict)
```
